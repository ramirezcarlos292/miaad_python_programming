# Ejercicio 1:
Análisis de la distribución de supervivencia por combinación de sexo y clase del pasajero.**


```python
# * Calcula la proporción de supervivencia para cada combinación de 'Sex' y 'Pclass'.
# * Identifica qué combinación tuvo la tasa de supervivencia más alta.
# * Identifica qué combinación tuvo la tasa de supervivencia más baja.
# como le hacemos? https://statsthinking21.github.io/statsthinking21-core-spanish-site/modeling-categorical-relationships.html
df['Pclass'] = df['Pclass'].astype(str)
df['Sex&Pclass'] = df['Sex'] + df['Pclass']

ej1 = df.copy()
ej1 = ej1[['Sex&Pclass', 'Survived']]
ej1 = ej1.groupby(by='Sex&Pclass').value_counts()

lst = pd.DataFrame(
    {"total_sex&pclass":
     [ej1['female1'].sum(), ej1['female1'].sum(),
      ej1['female2'].sum(), ej1['female2'].sum(),
      ej1['female3'].sum(), ej1['female3'].sum(),
      ej1['male1'].sum(), ej1['male1'].sum(),
      ej1['male2'].sum(), ej1['male2'].sum(),
      ej1['male3'].sum(), ej1['male3'].sum()]
     })

ej1
ej1 = ej1.reset_index()

ej1 = pd.concat([ej1, lst], axis=1)
ej1['Survival_rate_%'] = ej1['count'] / ej1['total_sex&pclass'] * 100

ej1 = ej1.loc[ej1['Survived'] == 1]
ej1
```

```python
print(f"La combinacion con tasa de supervivencia más alta fue: Pasajeros Mujeres de Primera Clase con 96.80% de supervivencia")
ej1.loc[ej1['Survival_rate_%'].idxmax()]
```

```python
print(f"La combinacion con tasa de supervivencia más baja fue: Pasajeros Hombres de Tercera Clase con 13.54% de supervivencia")
ej1.loc[ej1['Survival_rate_%'].idxmin()]
```

# Ejercicio 2:
Identificación de familias grandes a bordo.**

```python
# * Crea una nueva columna 'FamilySize' sumando las columnas 'SibSp' y 'Parch'.
# * Considera como "familia grande" a aquellas donde 'FamilySize' es mayor a 3.
# * Calcula el número de pasajeros en familias grandes.
# * Calcula la proporción de supervivencia entre los pasajeros que pertenecen a familias grandes.
ej2 = df.copy()
ej2['FamilySize'] = ej2['SibSp'] + ej2['Parch']
ej2['BigFamily'] = ej2['FamilySize'] > 3
ej2 = ej2[['Survived', 'BigFamily']]

ej2 = ej2.groupby(by='BigFamily').value_counts()
ej2 = ej2.reset_index()
ej2 = ej2.loc[ej2['BigFamily'] == True]
total = ej2['count'].sum()
ej2['Survival_rate'] = ej2['count'] / total * 100
print(f"Survival Rate for people who had big families a bordo fue de: 16.13%")
ej2['Survival_rate'].loc[ej2['Survived'] == 1]
```

# Ejercicio 3:
Segmentación por grupos de edad.

```python
# Clasifica a los pasajeros en las siguientes categorías de edad(tip puede resultar mas sencillo realizarlo con una función): menor de edad (< 18) y mayor de edad (>=18)
def under_age(value):
  if value < 18:
    return 'menor de edad'
  else:
    return 'mayor de edad'

ej3 = df.copy()
ej3['GroupAge'] = ej3['Age'].apply(under_age)
ej3['GroupAge'].value_counts().reset_index()
```

# Ejercicio 4:
Comparación entre promedios calculados manualmente y con Pandas.

```python
# Utiliza NumPy para calcular el promedio de las columnas 'Age' y 'Fare', ignorando valores nulos.
# Compara estos valores con los promedios obtenidos utilizando los métodos nativos de Pandas.
# Verifica que los resultados sean consistentes.
np_age_array = np.array(df['Age'])
np_age_array = np_age_array[~np.isnan(np_age_array)]
print(f"Numpy Age Mean: {float(np_age_array.mean()):.6f}")

np_fare_array = np.array(df['Fare'])
np_fare_array = np_fare_array[~np.isnan(np_fare_array)]
print(f"Numpy Age Fare: {float(np_fare_array.mean()):.6f}")

# Mean Age and Fare using pandas methods
pd.concat([df[['Age']].dropna().mean(), df[['Fare']].dropna().mean()])
```

# Ejercicio 5.
Creación de intervalos de clase usando NumPy y análisis con Pandas.

```python
# Divide la columna 'Fare' en 5 intervalos equidistantes utilizando la función numpy.linspace, el estudiante deberá investigar la operación de esta función en python.
# Crea una nueva columna en el DataFrame que asigne a cada pasajero el intervalo correspondiente de su tarifa.
# Calcula el número de pasajeros en cada intervalo utilizando Pandas y la proporción de supervivientes por intervalo.

ej5 = df.copy()
minn = min(ej5['Fare'])
maxx = max(ej5['Fare'])
intervalo = np.linspace(minn, maxx, num=5)

def agrupar_tarifa(fare):
  if fare <= intervalo[1]:
    return 'tarifa4'
  elif (fare > intervalo[1]) & (fare <= intervalo[2]):
    return 'tarifa3'
  elif (fare > intervalo[2]) & (fare <= intervalo[3]):
    return 'tarifa2'
  else:
    return 'tarifa1'

ej5['Fare_agrupado'] = ej5['Fare'].apply(agrupar_tarifa)
# Contar total de pasajeros de cada tarifa
s = ej5['Fare_agrupado'].value_counts()
s = s.reset_index()
s = s.rename(columns={'count':'total'})

# Contar solo los pasajeros sobrevivientes por tarifa
ej5 = ej5.loc[ej5['Survived'] == 1]
ej5 = ej5[['Survived', 'Fare_agrupado']]
ej5 = ej5.groupby(by='Fare_agrupado').value_counts()
ej5 = ej5.reset_index()

# Calcular la proporcion de sobrevivientes por intervalo
final = pd.merge(ej5, s, how='left', on='Fare_agrupado')
final['Survival_rate_%'] = final['count'] / final['total'] * 100

final
```
