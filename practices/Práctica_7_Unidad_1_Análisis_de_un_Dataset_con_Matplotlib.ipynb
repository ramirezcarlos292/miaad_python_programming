# 2. Crear los siguientes gráficos con Matplotlib:

## a) Gráfico de barras: <br>

* Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.

* Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

```python
'''
podemos ver como la especie setosa, tiende a ser mucho mas pequeña que las
otras variedades.
'''

a_species, means_length, means_width = [], [], []
for sp in species:
  a_species.append(sp)
  means_length.append(df['petal_length'].loc[df['species'] == sp].mean())
  means_width.append(df['petal_width'].loc[df['species'] == sp].mean())

bar_labels = ['blue', 'orange', 'green']
bar_colors = ['tab:blue', 'tab:orange', 'tab:green']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.bar(a_species, means_length, label=bar_labels, color=bar_colors)
ax1.set_title("Sepal Length by Species")
ax1.set_xlabel("Species")
ax1.set_ylabel("Sepal Length")

ax2.bar(a_species, means_width, label=bar_labels, color=bar_colors)
ax2.set_title("Sepal Width by Species")
ax2.set_xlabel("Species")
ax2.set_ylabel("Sepal Length")

plt.show()
```

## b) Histograma:

* Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).
* Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

```python
# Elegí el número de bins en 8, porque hace a cada uno los tres parecer una distribucion normal
maxx = max(df.sepal_length)
minn = min(df.sepal_length)
bins = np.linspace(minn, maxx, 8)
# print(bins)

plt.hist(df['sepal_length'].loc[df['species'] == 'setosa'], bins=bins, alpha=0.5, label='setosa', color='blue')
plt.hist(df['sepal_length'].loc[df['species'] == 'versicolor'], bins=bins, alpha=0.5, label='versicolor', color='orange')
plt.hist(df['sepal_length'].loc[df['species'] == 'virginica'], bins=bins, alpha=0.5, label='virginica', color='green')
plt.title('Histograma de Petal Length por especie')
plt.xlabel('Tamaño de petal length')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

## c) Gráfico de dispersión (scatter plot):

* Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.

* Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
# Como vimos en el primer ejercicio, setosa parece ser separable mas facilmente
# que sus otras dos variedades, y versicolor y virginica tienen una frontera más
# difusa para clasificar

def scatter_plot_prep(df, col1, col2):
  species = list(df.species.unique())
  colors = ['blue', 'orange', 'green']
  for idx, sp in enumerate(species):
    print(idx)
    x1 = df[col1].loc[df['species'] == sp]
    y1 = df[col2].loc[df['species'] == sp]
    plt.scatter(x1, y1, color=colors[idx])

  plt.title('Sepal Length vs Sepal Width Distribution by Species')
  plt.xlabel(col1)
  plt.ylabel(col2)
  plt.legend(species)
  plt.show()

scatter_plot_prep(df, 'petal_length', 'petal_width')
```

## d) Gráfico de cajas (box plot):

* Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).

* Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.

```python
def box_plot_prep(df, col):
  species = list(df.species.unique())

  data = {
      species[0]: list(df[col].loc[df['species'] == species[0]]),
      species[1]: list(df[col].loc[df['species'] == species[1]]),
      species[2]: list(df[col].loc[df['species'] == species[2]])
  }

  data_values = list(data.values())
  data_labels = list(data.keys())

  return data_values, data_labels

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
pl_values, pl_labels = box_plot_prep(df, 'sepal_length')
ax1.boxplot(pl_values, tick_labels=pl_labels, vert=True, patch_artist=True)
ax1.set_title("Box plot Sepal Length by Species")
ax1.set_xlabel("Species")
ax1.set_ylabel("Sepal Length")

pl_values, pl_labels = box_plot_prep(df, 'petal_length')
ax2.boxplot(pl_values, tick_labels=pl_labels, vert=True, patch_artist=True)
ax2.set_title("Box plot Petal Length by Species")
ax2.set_xlabel("Species")
ax2.set_ylabel("Petal Length")
```

## e) Gráfico de líneas:

* Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.

* Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
# en este último gráfico, podemos observar, al ordenar el dataset por variedad
# que los petal length son similares entre variedades, dando saltos en esta medición,
# marcados entre sí

index = np.arange(0, len(list(df.species)))
list_petal_length = list(df.petal_length)

plt.plot(index, list_petal_length, color='black')
plt.axvline(x=50, color='blue')
plt.axvline(x=100, color='orange')
plt.axvline(x=150, color='green')

plt.xlabel('Numero de muestra')
plt.ylabel('Petal length')
plt.title('Petal Length por especie')
```
