El archivo contiene varios correos electrónicos en donde se informa el valor de las acciones de diferentes empresas. El código deberá extraer el valor actual de las acciones que aparecen en cada correo electrónico e imprimirlo a pantalla. En este ejercicio debeás utilizar Expresiones Regulares

```python
# Comprobar que el archivo existe en drive
informe_acciones = '/content/drive/MyDrive/ClassFiles/informe_acciones.txt'
os.path.exists(informe_acciones)
```

```python
# Leer el archivo y buscar el valor de las acciones
with open(informe_acciones, 'r', encoding='latin-1') as archivo:
  valor_de_las_acciones = []
  for linea in archivo:
    linea = linea.rstrip()
    # Filtra todos los valores numericos
    datos_numericos = re.findall(r'\d+\.\d*', linea)
    if len(datos_numericos) > 0:
      valor_de_las_acciones.append(float(datos_numericos[0]))

valor_de_las_acciones
```
