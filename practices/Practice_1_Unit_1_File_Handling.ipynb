```python
# Saber si el archivo esta en el path correcto
print(os.path.exists('/content/drive/MyDrive/ClassFiles/mbox.txt'))
```

```python
# Leer el archivo
with open('/content/drive/MyDrive/ClassFiles/mbox.txt', 'r') as archivo:
  valores = []
  for linea in archivo:
    linea = linea.rstrip()
    # Si la linea no comienza con X-DSPAM-Confidence, ignorarla
    if not linea.startswith('X-DSPAM-Confidence'):
      continue
    # Si la linea comienza con X-DSPAM-Confidence, cortar los primeros 20 caracteres y convertir a flotante el numero y adjuntarlo a la lista
    valores.append(float(linea[20:]))

# Lista de valores extraidos
print(f"Valores de X_DSPAM-Confidence: \n {valores}")

# Cantidad de datos extraidos
cantidad_de_datos = len(valores)
print(f"Cantidad de valores extraidos: {cantidad_de_datos}")

# Sumatoria de los datos
sumatoria = 0
for valor in valores:
  sumatoria += valor
print(f"Sumatoria: {sumatoria}")

# Promedio de los datos
promedio = sumatoria/cantidad_de_datos
print(f"Promedio: {promedio}")

# Calcular Varianza
accum = 0
for valor in valores:
  accum += (valor - promedio)**2
var = accum / cantidad_de_datos
print(f"Varianza: {var}")
```

```python
# Comprobacion usando numpy
import numpy as np

# convertir lista de valores a numpy
valores = np.array(valores)

# Calcular sumatoria
print(f"Sumatoria: {valores.sum()}")

# Calcular promedio
print(f"Promedio: {valores.mean()}")

# Calcular la varianza
print(f"Varianza: {valores.var()}")
```
